```python
import pandas as pd
```

## Load

```python

# Load data
sports_df = pd.read_csv('olympics_data.csv')
sports_df.head()
```

# Clean

```python
# Checking to see count of datapoints
sports_df.count()
```

```python
# Dropping NaN
sports_df = sports_df.dropna(how='any')
sports_df.count()
```

## Female Sports

### Gold

```python
# Filtering for FEMALE and Medal (GOLD)
female_sports_df = sports_df.loc[(sports_df['Sex'] =='F') & (sports_df['Medal'] =='Gold') , :]
female_sports_df
```

```python
# Finding Age, Height, Weight Median

gold_female_sports_df = female_sports_df.groupby(['Event','Sport','Year', 'Country', 'NOC', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
gold_female_sports_df = gold_female_sports_df.median()
gold_female_sports_df[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
gold_female_sports_df.to_csv('female_bodycomp_age.csv', index=True)

```

## Female Sport Categories

```python
# Finding all female sports categories 
unique_f_sports_df = female_sports_df['Sport']
unique_f_sports_df = pd.DataFrame(unique_f_sports_df)

# unique_f_sports_df.Sport.unique()
```

## Female Sport Events

```python
# Finding all female sport events
unique_f_events_df = female_sports_df['Event']
unique_f_events_df = pd.DataFrame(unique_f_events_df)

# unique_f_events_df.Event.unique()
```

## Male Sports

### Gold

```python
# Filtering for MALE and  Medal (GOLD)

male_sports_df = sports_df.loc[(sports_df['Sex'] =='M') & (sports_df['Medal'] =='Gold'), :]
male_sports_df.head()
```

```python
# Finding Age, Height, Weight Median

gold_male_sports_df = male_sports_df.groupby(['Event','Sport','Year', 'Country', 'NOC', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
gold_male_sports_df = gold_male_sports_df.median()
gold_male_sports_df[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
gold_male_sports_df.to_csv('male_bodycomp_age.csv', index=True)

```

## Male Sport Categories

```python
# Finding all male sport categories 
unique_m_sports_df = male_sports_df['Sport']
unique_m_sports_df = pd.DataFrame(unique_m_sports_df)

# unique_m_sports_df.Sport.unique()
```

## Male Sport Events

```python
# Finding all male sport events
unique_m_events_df = male_sports_df['Event']
unique_m_events_df = pd.DataFrame(unique_m_events_df)

# unique_m_events_df.Event.unique()
```
