```python
import pandas as pd
import json
```

## Load

```python

# Load data
sports_df = pd.read_csv('olympics_data.csv')
sports_df.head()
```

# Clean

```python
# Checking to see count of datapoints
sports_df.count()
```

```python
# Dropping NaN
sports_df = sports_df.dropna(how='any')
sports_df.count()
```

## Female Sports

### Gold

```python
# Filtering for FEMALE and Medal (GOLD)
female_sports_df = sports_df.loc[(sports_df['Sex'] =='F') & (sports_df['Medal'] =='Gold') , :]
female_sports_df
```

```python
# Finding Age, Height, Weight Median

gold_female_sports_df = female_sports_df.groupby(['Event', 'Sport', 'Country', 'NOC', 'Year', 'Sex'])

# Find median for Age, Height, Weight
gold_female_sports_df = gold_female_sports_df.median()
gold_female_sports_df = gold_female_sports_df[["Age", "Height", "Weight"]]
gold_female_sports_df 
```

```python
# Reset index
gold_female_sports_df = gold_female_sports_df.reset_index(level=['Event', 'Sport', 'Country', 'NOC', 'Year', 'Sex'])
gold_female_sports_df
```

```python
# Setting Dataframe into JSON format
results = {}
for key, df_gb in gold_female_sports_df.groupby('NOC'):
    results[str(key)] = df_gb.to_dict('records')

# print(results)
```

```python
# Saving Dataframe into a JSON file 
with open('Test-HTML/static/js/female_bodycomp_age.json', 'w', encoding='utf-8') as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```

## Male Sports

### Gold

```python
# Filtering for MALE and Medal (GOLD)

male_sports_df = sports_df.loc[(sports_df['Sex'] =='M') & (sports_df['Medal'] =='Gold'), :]
male_sports_df.head()
```

```python
# Reset index
gold_male_sports_df = gold_male_sports_df.reset_index(level=['Event', 'Sport', 'Country', 'NOC', 'Year', 'Sex'])
gold_male_sports_df
```

```python
# Setting Dataframe into JSON format
results = {}
for key, df_gb in gold_male_sports_df.groupby('NOC'):
    results[str(key)] = df_gb.to_dict('records')

# print(results)
```

```python
# Saving Dataframe into a JSON file 
with open('Test-HTML/static/js/male_bodycomp_age.json', 'w', encoding='utf-8') as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```
