```python
import pandas as pd
```

```python
# Load data
gold_male_sports_df = pd.read_csv('Male-Sports/male_olympics_gold.csv')
silver_male_sports_df = pd.read_csv('Male-Sports/male_olympics_silver.csv')
bronze_male_sports_df = pd.read_csv('Male-Sports/male_olympics_bronze.csv')
```

```python
# Group by Event - Gold
gold_df_by_event = gold_male_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
gold_df_med_by_event = gold_df_by_event.median()
gold_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
gold_df_med_by_event.to_csv('Male-Events/olympics_male_gold_bodycomp_age.csv', index=True)
gold_df_med_by_event.to_json('Male-Events/olympics_male_gold_bodycomp_age.json')
```

```python
# Group by Event - Silver
silver_df_by_event = silver_male_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
silver_df_med_by_event = silver_df_by_event.median()
silver_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
silver_df_med_by_event.to_csv('Male-Events/olympics_male_silver_bodycomp_age.csv', index=True)
silver_df_med_by_event.to_json('Male-Events/olympics_male_silver_bodycomp_age.json')
```

```python
# Group by Event - Bronze
bronze_df_by_event = bronze_male_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
bronze_df_med_by_event = bronze_df_by_event.median()
bronze_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
bronze_df_med_by_event.to_csv('Male-Events/olympics_male_bronze_bodycomp_age.csv', index=True)
bronze_df_med_by_event.to_json('Male-Events/olympics_male_bronze_bodycomp_age.json')
```
