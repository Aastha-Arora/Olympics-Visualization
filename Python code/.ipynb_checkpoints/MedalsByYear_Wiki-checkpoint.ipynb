## Scraping Wikipedia for Summer Olympics medals table

### 1980
---

```python
year = 1980

# Query URL to scrape
query_url = (f'https://en.wikipedia.org/wiki/{year}_Summer_Olympics_medal_table')

print(query_url)
    
tables = pd.read_html(query_url,index_col = None)
    
medals_table_1980 = tables[0]

# Dropping the last row as its irrelevant
medals_table_1980.drop(medals_table_1980.tail(1).index,inplace=True)

# Adding 'Year' column
medals_table_1980['Year'] = year

medals_table_1980.rename(columns={'Total': 'Medals'},inplace = True)

# Rearranging columns
cols = medals_table_1980.columns.tolist()

cols = cols[-1:] + cols[:-1]

medals_table_1980 = medals_table_1980[cols]

medals_table_1980.head(10)
```

### Concatenating medals data of all the years
---

```python
df_list = [medals_table_1980,medals_table_1984,medals_table_1988,
           medals_table_1992,medals_table_1996,medals_table_2000,
           medals_table_2004,medals_table_2008,medals_table_2012,
           medals_table_2016]

medals_by_year = pd.concat(df_list, axis=0)

medals_by_year = medals_by_year.sort_values(['Year','Medals','Nation'], ascending = [True, False, True])

medals_by_year = medals_by_year.reset_index(drop=True)

medals_by_year.head(20)
```
