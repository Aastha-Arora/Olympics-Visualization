# Summer Olympics Medal Tally 1980-2016
---

```python
import pandas as pd
```

```python
Olympic_data = pd.read_csv('../static/assets/data/athlete_events.csv')
Olympic_data.head(10)
```

```python
# Selecting only required columns
filtered_df = Olympic_data[['NOC','Year','Season','Event','Medal']]

# Dropping NaN values in 'Medal' column as we are inerested in only medal count
filtered_df = filtered_df.dropna()

# Filtering the dataFrame for only summer olympics
filtered_df = filtered_df[filtered_df['Season'] == 'Summer']

# Filtering the dataFrame for year >= 1980
filtered_df = filtered_df[filtered_df['Year'] >= 1980]

# Drop duplicate rows for team events.
filtered_df = filtered_df.drop_duplicates(['NOC','Year','Season','Event','Medal'])

filtered_df.head(10)
```

```python
filtered_df.shape
```

```python
# Selecting required columns
filtered_df = filtered_df[['NOC','Year','Medal']]

filtered_df.head(10)
```

```python
medals_by_year = filtered_df.groupby(['Year','NOC'], sort = False).count()['Medal']

medals_by_year = pd.DataFrame(medals_by_year).rename(columns={'Medal': 'Medals'}).reset_index()

medals_by_year.head(10)
```

```python
medals_by_year = medals_by_year.sort_values(['Year','NOC','Medals'], ascending = [True,True,False])

# Using the drop parameter to avoid the old index being added as a column:
medals_by_year = medals_by_year.reset_index(drop = True)

medals_by_year.head(10)
```

### Adding Country name based on NOC from noc_regions.csv

```python
# Read the noc_regions csv file to get Nation name
noc_regions = pd.read_csv('../static/assets/data/noc_regions.csv')
noc_regions.head(10)
```

```python
medals_by_year = medals_by_year.merge(noc_regions, how ='inner')
medals_by_year.head(10)
```

```python
# Rename 'region' column to 'Nation'
medals_by_year = medals_by_year.rename(columns = {'region':'Nation'})

# Selecting only required columns
medals_by_year = medals_by_year[['Year','Nation','Medals']]

# Sorting the dataframe by year(ascending), Medals(descending) and Nation (alphabetically)
medals_by_year = medals_by_year.sort_values(['Year','Medals','Nation'], ascending = [True,False,True])

medals_by_year = medals_by_year.reset_index(drop=True)

medals_by_year.head(10)
```

## Part 2: Finding Gold, Silver and Bronze medal count
---

### Gold Count

```python
# Filter the dataFrame for Gold medals
gold_medals_df = filtered_df[filtered_df['Medal'] == 'Gold']

# Rename 'Medal' column as 'Gold'
gold_medals_df = gold_medals_df.rename(columns={'Medal': 'Gold'})

# Sort the dataframe by Year
gold_medals_df = gold_medals_df.sort_values(['Year'])

# Groupby 'Year' and 'NOC' and find count of Gold medals
gold_medals_df = gold_medals_df.groupby(['Year','NOC'], sort = False).count()['Gold']

gold_medals_df = pd.DataFrame(gold_medals_df)

# Sort the dataframe by 'Year'(ascending) and 'Gold'(descending)
gold_medals_df = gold_medals_df.sort_values(['Year','Gold','NOC'], ascending = [True, False,True])

gold_medals_df.reset_index(inplace=True)

gold_medals_df.head(10)
```

### Silver count

```python
# Filter the dataFrame for Silver medals
silver_medals_df = filtered_df[filtered_df['Medal'] == 'Silver']

# Rename 'Medal' column as 'Silver'
silver_medals_df = silver_medals_df.rename(columns={'Medal': 'Silver'})

# Sort the dataframe by Year
silver_medals_df = silver_medals_df.sort_values(['Year'])

# Groupby 'Year' and 'NOC' and find count of Silver medals
silver_medals_df = silver_medals_df.groupby(['Year','NOC'], sort = False).count()['Silver']

silver_medals_df = pd.DataFrame(silver_medals_df)

# Sort the dataframe by 'Year'(ascending) and 'Silver'(descending)
silver_medals_df = silver_medals_df.sort_values(['Year','Silver'], ascending = [True,False])

silver_medals_df.reset_index(inplace=True)

silver_medals_df.head(10)
```

### Bronze Count

```python
# Filter the dataFrame for Bronze medals
bronze_medals_df = filtered_df[filtered_df['Medal'] == 'Bronze']

# Rename 'Medal' column as 'Bronze'
bronze_medals_df = bronze_medals_df.rename(columns={'Medal': 'Bronze'})

# Sort the dataframe by Year
bronze_medals_df = bronze_medals_df.sort_values(['Year'])

# Groupby 'Year' and 'NOC' and find count of Bronze medals
bronze_medals_df = bronze_medals_df.groupby(['Year','NOC'], sort = False).count()['Bronze']

bronze_medals_df = pd.DataFrame(bronze_medals_df)

# Sort the dataframe by 'Year'(ascending) and 'Silver'(descending)
bronze_medals_df = bronze_medals_df.sort_values(['Year','Bronze'], ascending = [True,False])

bronze_medals_df.reset_index(inplace=True)

bronze_medals_df.head(10)
```

### Merging Gold, Silver and Bronze medals
---

```python
# Merge Gold and Silver

gold_silver_df = gold_medals_df.merge(silver_medals_df, how='outer').fillna(0)

gold_silver_df.head(10)
```

```python
# Merge Bronze to Gold and Silver

all_medals_df = gold_silver_df.merge(bronze_medals_df, how='outer').fillna(0)

all_medals_df.head(10)
```

```python
all_medals_df['Medals'] = all_medals_df['Gold'] + all_medals_df['Silver'] + all_medals_df['Bronze']

all_medals_df = all_medals_df.sort_values(['Year','Medals','NOC'], ascending = [True,False,True])

all_medals_df = all_medals_df.reset_index(drop=True)

all_medals_df.head(10)
```

### Adding Country name based on NOC from noc_regions.csv

```python
noc_regions = pd.read_csv('../static/assets/data/noc_regions.csv')
noc_regions.head(10)
```

```python
all_medals_df = all_medals_df.merge(noc_regions, how ='inner')
all_medals_df.head(10)
```

```python
all_medals_df = all_medals_df.rename(columns = {'region':'Nation'})

all_medals_df = all_medals_df[['Year','Nation','NOC','Gold','Silver','Bronze','Medals']]

all_medals_df.head(10)
```
